```
import mtcnn
```

#Importando os métodos necessários


```
from mtcnn.mtcnn import MTCNN
from cv2 import imread
```

# Carregando a imagem de teste

```
img_array=imread('/content/drive/My Drive/Colab Notebooks/AngelinaJolie.jpeg')
```

# Instanciando o detector

```
face_detector=MTCNN()
```

# Realizando a detecção da face

```
faces=face_detector.detect_faces(img_array)
```

#Verificando o tipo do objeto


```
type(faces)
```

#Listando as faces detectadas

```
for face in faces:
  print(face)
```

# Verificando as bounding boxes detectadas

```
for face in faces:
  print(face['box'])
```

# Verificando os demais atributos

```
for face in faces:
  print('Confiança:{}'.format(face['confidence']))
  print('Pontos chave:{}'.format(face['keypoints']))
```

# Plotando um retângulo na face detectada

```
#Importando o método responsável para desenhar o retângulo
```

```
from cv2 import rectangle, circle
from cv2 import imshow, waitKey, destroyAllWindows
```

```
for face in faces:
  x,y,width,height=face['box']
  x2,y2=x+width,y+height
  # Desenhando o retângulo na face detectada
  rectangle(img_array, (x,y), (x2,y2), (0,0,255), 3)
  # Desenhando um circulo em cada ponto da face detectada
for ponto, valor in face['keypoints'].items():
  circle(img_array,valor,radius=8,color=(255,0,0))
```

# Exibindo a imagem

```
cv2_imshow(img_array)
waitKey(0)
destroyAllWindows()
```
